```python
from bs4 import BeautifulSoup as BS
```

```python
addGeneIDS = ['45978', '45980', '45985', '45989', '38207', '38208', '38209', '44449', '44451', '44452', '44453', '44456']
```

```python
def getUrl(url):
    resp = r.get(url)
    if resp.ok:
        return resp.content
    else:
        raise Exception(url) #TODO
```

```python
addGeneUrls = ["https://www.addgene.org/{}/".format(ID) for ID in addGeneIDS]
```

```python
def getSeqPrimer(soup, start):
    try:
        return [x for x in soup.find_all(lambda tag: tag.name=="h2" and "Cloning" in tag.text)[0].parent.find_all(
        lambda tag: tag.name=="span" and "sequencing" in tag.text) if start in x.text][0].next.next.strip()
    except IndexError:
        return ""
```

```python
def getCloningMethod(soup):
    try:
        return soup.find_all(lambda tag: tag.name=="h2" and "Cloning" in tag.text)[0].parent.find_all(
        lambda tag: tag.name=="span" and "method" in tag.text)[0].next.next.strip()
    except IndexError:
        return ""
```

```python
df = pd.read_csv("template.csv")
genBankFiles = {}
for url in addGeneUrls:
    html = getUrl(url)
    print(url)
    soup = BS(html, "html.parser")
    data = {"gene_name_short": soup.find_all("span", class_="material-name")[0].string,
            "gene_name_long": "",
            "product": "",
            "plasmid_name": "",
            "backbone": soup.find(id="detail-sections").find_all("li")[0].text.split("\n")[3].strip(),
            "sequencing_primer_forward": getSeqPrimer(soup, "5"),
            "sequencing_primer_reverse": getSeqPrimer(soup, "3"),
            "genbank_source_id": "",
            "genbank_source_link": "",
            "genbank_protein_id": "",
            "genbank_protein_link": "",
            "locus_tag": "",
            "uniprot_link": "",
            "origin_organism": "",
            "target_organism": "",
            "description": "Grown at {}".format(soup.find(id="detail-sections").find_all("section")[2]
                                                .find_all("div")[2].next.next.strip()),
            "growth_antibotic": soup.find(id="detail-sections").find_all("section")[2].find_all("div")[0]
            .next.next.strip(),
            "shipping_strain": soup.find(id="detail-sections").find_all("section")[2].find_all("div")[2]
            .next.next.strip(),
            "expression_strain": "",
            "expression_promoter": "",
            "expression_inducer": "",
            "cloning_method": getCloningMethod(soup),
            "cloning_enzyme": "",
            "cloning_part_type": "",
            "cloning_prefix": "",
            "cloning_suffix": "",
            "insert_sequence": "",
            "construct_insert_size": "",
            "ori": "",
            "copy_number": soup.find(id="detail-sections").find_all("section")[2].find_all(
                lambda tag: tag.name=="div" and "Copy" in tag.text)[0].next.next.strip(),
            "bsl": "BSL1",
            "export_controls": "None",
            "safety_other_information": "",
            "attribution": ", ".join([x.text.strip() for x in soup.find(id="plasmid-description-list").find_all(
            lambda tag: (tag.name == "a") and ("href" in tag.attrs) and ("/browse/pi/" in tag["href"]))]),
            "citations": " ".join([a.strip() for a in soup.find_all("cite")[-1]
                                   .text.split("\n")]).strip() if len(soup.find_all("cite"))>0 else "",
            "canary_notice": "we are unaware of any 3rd party property rights on the use of this item"}
    
    html = getUrl(url+"sequences/")
    newSoup = BS(html, "html.parser")
    genbankLink = newSoup.find_all("a", class_="genbank-file-download")[0]["href"]
    
    name = data["gene_name_short"]
    
    genBankFiles[name] = io.BytesIO(getUrl(genbankLink))
    
    df = df.append(data, ignore_index=True)
df

```

```python
import zipfile

zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "a", zipfile.ZIP_DEFLATED, False) as zip_file:
    for file in genBankFiles:
        zip_file.writestr(file+".gb", genBankFiles[file].getvalue())
```
